```python
import numpy as np
```

```python
figure = plt.figure()
axes = figure.add_subplot(111)

x = [1,2,3,4]
y = [2,3,5,9]

x2 = np.array([1,2,3,4])
y2 = np.array([4,1,3,1])

axes.plot(x,y, color='red', linestyle='dotted', marker='o')
axes.plot(x2,y2, linestyle='dashed', marker='^')
```

```python
figure = plt.figure()
axes = figure.add_subplot(111)

x = [1,2,3,4]
y = [2,3,5,9]

x2 = [2,4,6,8]
y2 = [4,4,3,6]

axes.bar(x,y, color='cyan')
axes.bar(x2,y2, color='whitesmoke')

plt.show()
```

```python
figure = plt.figure()
axes = figure.add_subplot(111)
axes2 = axes.twinx()

x = [1,2,3,4]
y = [2,3,5,9]

x2 = [1,2,3,4]
y2 = [4,6,1,9]

axes.bar(x,y, color='cyan', label='bar')
axes2.plot(x2,y2, color='darkgreen', label = 'plot')

axes.legend()
axes2.legend(loc=3)

plt.show()
```

```python
figure = plt.figure()
axes = figure.add_subplot(111)
axes2 = axes.twinx()

x = [1,2,3,4]
y = [2,3,5,9]

x2 = [1,2,3,4]
y2 = [4,6,1,9]

axes.scatter(x,y, color='cyan')
axes2.scatter(x2,y2, color='darkgreen')

plt.show()
```
